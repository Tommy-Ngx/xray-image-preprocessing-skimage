# Read preprocessed images and save as csv

## Setup

```python
from glob import iglob
import h5py
import skimage.io as io
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from skimage.util import img_as_ubyte
```

```python
# path of preprocessed train images
path_train_preprocessed = "boneage-training-dataset/train_preprocessed/"
path_train_csv = "boneage-training-dataset/train_csv/"

path_test_preprocessed = "boneage-training-dataset/train_preprocessed/"
path_test_csv = "boneage-training-dataset/train_csv/"
```

## Read the images, flatten them and save them in a dataframe

```python
i=0
labels = []
hands = pd.DataFrame([])

for path in iglob(path_train_preprocessed + '9*.png'):
    img=io.imread(path) 
    img = img_as_ubyte(img)
    img=img/255.0
    path_split= path.split("/")
    label = path_split[-1].split('.')[0]
    labels.append(label)
    hand = pd.Series(img.flatten(),name=path)

    hands = hands.append(hand)
```

```python
hands.head()
```

```python
# add the label to the dataframe
hands["id"] = labels
hands.head()
```

```python
hands.to_csv(path_train_csv + 'train9_sample.csv')
```
